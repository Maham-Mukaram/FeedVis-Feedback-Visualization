```python
import pickle

product_asin='B018Y229OU'

with open(f'Reviews_{product_asin}.pkl', 'rb') as f:
    data = pickle.load(f)
```

```python
import matplotlib.pyplot as plt
import numpy as np
year = 2016
predef_y = np.array([len(data.predef_sentiment[(data.predef_sentiment=='positive')&(data.year==year)]),
                     len(data.predef_sentiment[(data.predef_sentiment=='negative')&(data.year==year)]),
                     len(data.predef_sentiment[(data.predef_sentiment=='neutral')&(data.year==year)])])
mylabels = ["Positive", "Negative", "Neutral"]
mycolors = ["green", "red", "grey"]

print(predef_y)
plt.pie(predef_y, labels = mylabels, colors = mycolors)
plt.show() 
```

```python
zc_y = np.array([len(data.zc_sentiment[data.zc_sentiment=='positive']),
                     len(data.zc_sentiment[data.zc_sentiment=='negative']),
                     len(data.zc_sentiment[data.zc_sentiment=='neutral'])])

print(zc_y)
plt.pie(zc_y, labels = mylabels, colors = mycolors)
plt.show() 
```
